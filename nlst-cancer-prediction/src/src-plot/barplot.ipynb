```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the one model auc values

categories = ['Baseline', 
#              'DeltaT1', 
              'Baseline+DeltaT1', 'Baseline+DeltaT2' ]

models = ['Non-Combat', 'Combat-Thickness Category', 'Combat-Kernel Type']

# AUC values for each category and model
auc_values = {
    'Baseline': [0.6238095, 0.6328822, 0.6538847],
#    'DeltaT1': [ ],
    'Baseline+DeltaT1': [0.6535338 , 0.6434586, 0.6186466 ],
   'Baseline+DeltaT2': [0.7289223 , 0.7268672 , 0.7068672 ]
}
# Confidence intervals (95%) for each category and model
sd_values = {
    'Baseline': [0.014711040665330225, 0.028900265301066014 ,0.02466951804407619],
#    'DeltaT1': [],
   'Baseline+DeltaT1': [ 0.04555224962587469, 0.03987071285233063 , 0.07760005757888971 ],
   'Baseline+DeltaT2': [ 0.02687243723265535, 0.028229481524108806 ,0.02722731100086823 ]
}



# Set the width of the bars
bar_width = 0.35

# Create an array of indices for the x-axis
x = np.arange(len(categories))





# Setup the figure and axes
fig, ax = plt.subplots(figsize=(12, 8))

# Calculate positions for each model within each category
num_models = len(models)
total_width = num_models * bar_width
spacing = 0.2  # Spacing between groups of bars

# X positions of the categories
x = np.arange(len(categories)) * (total_width + spacing)

# Plot the bars for each model
for i, model in enumerate(models):
    # Calculate offset from base x positions
    offsets = x + i * bar_width
    auc_data = [auc_values[category][i] for category in categories]
    sd_data = [sd_values[category][i] for category in categories]

    # Adding the bars for each category per model
    ax.bar(offsets, auc_data, bar_width, label=model, yerr=sd_data, capsize=5)

# Labeling and aesthetics
ax.set_xlabel('Category')
ax.set_ylabel('AUC')
ax.set_title('AUC by Model and Category with SD as Error Bars')
ax.set_xticks(x + total_width / 2 - bar_width / 2)  # Adjusting x-ticks to center them under the group
ax.set_xticklabels(categories)
ax.legend()


# Set labels and title
ax.set_xlabel('Feature sets', fontsize=14, labelpad=20)
ax.set_ylabel('AUC', fontsize=14, labelpad=20)
ax.set_title('The Median of AUC on test dataset over feature selection methods for Lung Cancer risk prediction using Decision Tree Model', pad=20)
ax.set_xticks(x + bar_width / 2)
ax.set_xticklabels(categories)
ax.legend()
# Set a custom y-axis limit
ax.set_ylim(0, 0.9)  # Adjust the limits as needed

# Adjust layout
#plt.tight_layout()

plt.savefig("/home/ulaval.ca/lesee/projects/Project-NLST/results/Plots/auc-DT.png")
# Display the plot
plt.show()
```
