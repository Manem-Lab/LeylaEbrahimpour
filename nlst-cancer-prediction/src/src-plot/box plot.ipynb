```python
import pandas as pd

excel_file = '/home/ulaval.ca/lesee/projects/Project-NLST/results/ML-NLST-results.xlsx'

# Get names of all sheets
xls = pd.ExcelFile(excel_file)
sheet_names = xls.sheet_names[0:3]
print(sheet_names)


# Load each sheet into a dictionary of DataFrames
dataframes = {sheet: pd.read_excel(excel_file, sheet_name=sheet) for sheet in sheet_names}
```

```python
sheet_names = ['T0', 'delta', 'T0+delta']
```

```python
for sheet_name, df in dataframes.items():
    # Check for missing values and handle them
    df.dropna(inplace=True)  # Example: drop missing values
#    print(df)
    
    # Convert data types if necessary
    df['AUC_test'] = df['AUC_test'].astype(float)
    df['Model'] = df['Model'].astype('category')
    df['Category'] = df['Category'].astype('category')
    df['Selection_Method'] = df['Selection_Method'].astype('category')
#    agg_df = df.groupby(['Model', 'Category'])['AUC_test'].median().reset_index()
#    print(agg_df)

    # Replace the DataFrame in the dictionary
    dataframes[sheet_name] = df
    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# Set the style
sns.set(style="whitegrid")

# Loop through the DataFrames and plot each one
for sheet_name, df in dataframes.items():
#    print(df)
    plt.figure(figsize=(20, 10))  # Set the figure size
    boxplot = sns.boxplot(x='Model', y='AUC_test', hue='Category', data=df, palette='Set2',  
                          width=0.6,  # Adjust width to make boxes thinner
                          dodge=True  # Ensure categories within each model are grouped together but separate
                         )
    plt.legend(title=None, fontsize=20, loc= "lower right")
    models = df['Model'].unique()
    n_models = len(models)
    for i in range(1, n_models):
        plt.axvline(x=i - 0.5, color='grey', linestyle='--')  # x position is b
#    plt.title(f'Median AUC by ML Model and Category Across All Feature Methods for {sheet_name} features', fontsize=20)
    plt.ylabel('AUC', labelpad = 25, fontsize=25)
    plt.xlabel('ML Models', labelpad = 25, fontsize=25)
    plt.xticks(fontsize=20)  # Rotate model names for better visibility
    plt.yticks(fontsize=20)  # Rotate model names for better visibility
    
    # Configure major and minor ticks on the y-axis
    ax = plt.gca()
    ax.yaxis.set_major_locator(MultipleLocator(0.05))  # Major ticks every 0.05
    ax.yaxis.set_minor_locator(MultipleLocator(0.01))  # Minor ticks every 0.01

    # Adjust horizontal grid lines - make them lighter
    ax.grid(True, 'major', 'y', linestyle='-', linewidth=0.5, color='grey', alpha=0.9)  # Major grid lines
    ax.grid(True, 'minor', 'y', linestyle='-', linewidth=0.25, color='grey', alpha=0.7)  # Minor grid lines, more subtle
    
    plt.savefig('boxPlot_'+ sheet_name+'_.png', dpi=300, bbox_inches='tight')  # Save as PNG with high resolution
    plt.show()
```
