```python
import pandas as pd
import numpy as np
# Path to your Excel file
file_path = '/home/ulaval.ca/lesee/projects/Project-NLST/results/ML-NLST-results.xlsx'
xl = pd.ExcelFile(file_path)
sheets_names = xl.sheet_names[0:3]
# Load each sheet into a separate DataFrame
sheet1 = pd.read_excel(file_path, sheet_name=sheets_names[0])
sheet2 = pd.read_excel(file_path, sheet_name=sheets_names[1])
sheet3 = pd.read_excel(file_path, sheet_name=sheets_names[2])
#sheet4 = pd.read_excel(file_path, sheet_name=sheets_names[3])

#sheets = [sheet1, sheet2, sheet3, sheet4]
sheets = [sheet1, sheet2, sheet3]
```

```python
print(sheets_names)
```

```python
# Function to get maximum AUC_test values for each category
def max_auc_per_category(df):
    return df.groupby('Category')['AUC_test'].max()

# Apply this function to each sheet
max_aucs = [max_auc_per_category(sheet) for sheet in sheets]

# Combine the results into a single DataFrame
max_aucs_df = pd.concat(max_aucs, axis=1)
max_aucs_df.columns = sheets_names
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Prepare the figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Number of categories
categories = max_aucs_df.index
n_categories = len(categories)

# Width of the bars
bar_width = 0.25

# Adjust the inter-group spacing
group_spacing = 0.3  # Adjust space between groups
indices = np.arange(n_categories) * (1 + group_spacing)

# Define colors (pastel shades)
#colors = ['#ffd1dc', '#b4f8c8', '#fde4cf', '#a0e7e5']
colors = plt.cm.Pastel2(np.linspace(0, 1, len(sheets)))  # Generate distinct pastel colors
#colors = ['#76B7B2', '#AF7AA1', '#EDC948']
#colors = plt.cm.viridis(np.linspace(0, 1, len(sheets)))  # generate distinct colors

# Plot each set of bars
#for i, (sheet_name, values) in enumerate(max_aucs_df.items()):
#    ax.bar(indices + i * bar_width, values, width=bar_width, label=sheet_name, color=colors[i])

# Plot each set of bars
for i, (sheet_name, values) in enumerate(max_aucs_df.items()):
    ax.bar(indices + i * bar_width - bar_width * 0.5, values, width=bar_width, label=sheet_name, color=colors[i % len(colors)], edgecolor='black')

# Find the overall maximum AUC to use for the horizontal line
overall_max_auc = max_aucs_df.max().max()

# Draw a dashed horizontal line at the overall maximum AUC
ax.axhline(y=overall_max_auc, color='red', linestyle='--', linewidth=2)

# Annotate the line to show its value
ax.text(x=-0.5, y=overall_max_auc, s=f'Max AUC: {overall_max_auc:.2f}', color='red', va='bottom')


# Set axis labels and chart title
ax.set_ylabel('AUC of the best model', fontsize=25, labelpad =20)
#ax.set_title('Comparison of the AUC for the best model within each pipeline using different feature sets')

# Set x-ticks to be in the middle of the groups
ax.set_xticks(indices)
ax.set_xticklabels(categories, fontsize=20)
ax.tick_params(axis='y', labelsize=20)  # Adjust the y-tick label size


# Adding a legend and place it outside of the plot
ax.legend(title=None, loc='upper left', bbox_to_anchor=(1, 1), fontsize=20)

# Show plot
plt.xticks(rotation=45)
#from matplotlib.ticker import FormatStrFormatter
#plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.tight_layout()
plt.savefig('max_auc_comparison_by_pipelines_noT2_pastel.png', dpi=300, bbox_inches='tight')  # Save as PNG with high resolution
plt.show()
```
