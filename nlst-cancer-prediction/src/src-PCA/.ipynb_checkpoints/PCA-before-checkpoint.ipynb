```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import os
```

```python
# Load data from Excel file
main_dir= "/home/ulaval.ca/lesee/projects/Project-NLST/"

df_cohort1_2_rads = pd.read_csv(os.path.join(main_dir,'data/radiomicsfeatures_kheops-NLST-Dmitrii-Cohort1_2_Laptop_v2.csv'))

df_cohort_match = pd.read_csv(os.path.join(main_dir,'data/combat_params_cohort1_2.csv'))


df_cohort1_2_rads_combat = pd.merge(df_cohort1_2_rads, df_cohort_match, on='SeriesInstanceUID', how='inner')        
```

```python
#print(list(df_cohort1_2_rads_combat))
```

```python
combat_cols = ['Manufacturer', 'ModelName', 'ConvolutionKernel',
            'ReconDiameter', 'SliceThickness', 'kVp', 'mAs', 'PixelSpacing']
data_combat = df_cohort1_2_rads_combat[combat_cols]

data_features = df_cohort1_2_rads_combat.drop(columns=['Unnamed: 0', 'PatientID_x', 'StudyDate_x', 'SeriesDescription_x',
                                                       'Kernel', 'CtSlices', 'SegSlices', 'SeriesInstanceUID','PatientID_y',
                                                       'StudyDate_y', 'StudyInstanceUID', 'SeriesDescription_y', 'ImageType', 
                                                       'Manufacturer', 'ModelName', 'ConvolutionKernel', 'ReconDiameter',
                                                       'SliceThickness', 'kVp', 'mAs', 'PixelSpacing'])
```

```python
value_counts = data_combat['Manufacturer'].value_counts()
# Identify values that occur exactly once
print(value_counts.index)
```

```python
# Define the custom function to categorize the kernel types
def categorize_kernel(kernel):
#    sharp_kernels = ['BONE', 'B80f', 'B60f', 'B50f', 'B30f', 'LUNG', 'B70f']
#    soft_kernels = ['STANDARD', 'C', 'D', 'B', 'FC51', 'FC10', 'FC30', 'FC02', 'FC01', 'A', 'FC50', 'FC82']
    sharp_kernels = ['D', 'B70f', 'B80f', 'BONE', 'FC82', 'LUNG', 'B60f', 'FC51', 'FC30', 'FC50']
    soft_kernels = ['B30f', 'B30s', 'FC10', 'FC02', 'FC01', 'A',  'STANDARD', 'B', 'C', 'B50s', 'B50f']
 #    sharp_kernels = ['LUNG', 'D', 'B50s', 'B60f', 'B70f', 'B80f', 'B50f', 'BONE' ]
#    soft_kernels = ['STANDARD', 'C', 'B30f', 'B30s', 'B', 'A']    
    
    
    if kernel in sharp_kernels:
        return 'Sharp'
    elif kernel in soft_kernels:
        return 'Soft'
    else:
        return 'Unknown'  # For any values not explicitly listed
    
# Define the function for categorization
def categorize_thickness(thickness):
    if thickness < 2.0:
        return '1.0-2.0 (mm)'
    elif thickness <= 3.0:
        return '2.0-3.0 (mm)'
    else:
        return '3.0-4.0 (mm)'

# Apply the function to create a new column
data_combat['ThicknessCategory'] = data_combat['SliceThickness'].apply(categorize_thickness)


# Apply the function to the original column to create a new one
data_combat['KernelType'] = data_combat['ConvolutionKernel'].apply(categorize_kernel)

#print(data_combat)
```

```python
# Remove the first row (index 0)
#data_combat = data_combat.iloc[1:]

# Separate the slice thicknesses from the features
batch_TC = data_combat['ThicknessCategory']
batch_KT = data_combat['KernelType']
batch_man = data_combat['Manufacturer']
batch_kernel = data_combat['ConvolutionKernel']
#print(slice_thicknesses)

#print(features)
                        
```

```python
# Assuming data_features, batch, and kernel data are already defined
scaler = StandardScaler()
standardized_features = scaler.fit_transform(data_features)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(standardized_features)

pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
pca_df['ThicknessCategory'] = batch_TC  # Assume 'batch_TC' contains thickness categories
pca_df['KernelType'] = batch_KT       # Assume 'batch_KT' contains kernel type data
pca_df['Manufacturer'] = batch_man       # Assume 'batch_KT' contains kernel type data
pca_df['ConvolutionKernel'] = batch_kernel       # Assume 'batch_KT' contains kernel type data

# Define custom colors with high contrast
custom_colors_thickness = {
    '1.0-2.0 (mm)': '#FF6347',  # Tomato red
    '2.0-3.0 (mm)': '#4682B4',  # Steel blue
    '3.0-4.0 (mm)': '#32CD32'   # Lime green
}
custom_colors_kernel = {
    'Sharp': '#FFD700',      # Gold
    'Soft': '#6A5ACD',       # Slate blue
    'Unknown': '#8B4513'     # Saddle brown
}

custom_colors_man = {
    'GE MEDICAL SYSTEMS': '#FFD700',      # Gold
    'SIEMENS': '#6A5ACD',       # Slate blue
    'Philips': '#8B4513',     # Saddle brown
    'TOSHIBA': '#FF6347'  # Tomato red
}
def parse_range(text):
    """Parse the numeric part of the range from text and return the starting value as a float."""
    # This extracts the numeric value before the hyphen and removes the "mm"
    return float(text.split('-')[0].strip())

# Define the plotting function
def plot_pca_by_category(df, category, ax, title, colors, show_y_labels=True, point_size=50):
#    unique_categories = sorted(df[category].unique())  # Sort categories
#    colors = {cat: plt.cm.jet(i / len(unique_categories)) for i, cat in enumerate(unique_categories)}
    category_counts = df[category].value_counts()
    # Sort categories from most to least populated
    sorted_categories = category_counts.sort_values(ascending=False).index.tolist()
    
    for label in sorted_categories:
        subset = df[df[category] == label]
        ax.scatter(subset['PC1'], subset['PC2'], c=[colors[label]], label=label, alpha=0.7, s=point_size)

    # Manually sort the handles and labels for the legend from least to most populated
    if category == 'ThicknessCategory':
        handles, labels = ax.get_legend_handles_labels()
        labels_handles = sorted(zip(labels, handles), key=lambda t: parse_range(t[0]))
        labels, handles = zip(*labels_handles)  # Unzip the sorted pairs
        ax.legend(handles, labels, fontsize=20)
    else:
        ax.legend(fontsize=20) 


    ax.set_title(title, fontsize=25, pad = 20)
    ax.set_xlabel('Principal Component 1 (PC1)', labelpad = 25, fontsize=20)
    ax.set_ylabel('Principal Component 2 (PC2)', labelpad = 25, fontsize=20)
#    ax.legend(handles, labels, fontsize=20)
    ax.grid(True)  # Enable grid
    # Set tick label size
    ax.tick_params(axis='both', which='major', labelsize=20)  # Increase tick label size
    if not show_y_labels:
        ax.set_yticklabels([])  # Hide y-axis labels for the second plot
    

# Create a figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))  # 1 row, 2 columns

# Plot PCA for ThicknessCategory
plot_pca_by_category(pca_df, 'ThicknessCategory', ax1, 'PCA by Thickness Category', custom_colors_thickness)
ax1.set_ylabel('Principal Component 2 (PC2)')  # Set y-label for the first subplot

# Plot PCA for KernelType
plot_pca_by_category(pca_df, 'KernelType', ax2, 'PCA by Kernel Type', custom_colors_kernel, show_y_labels=False)
ax2.set_ylabel('')  # Remove y-label for the second subplot to avoid redundancy

# Plot PCA for Manufacturer
#plot_pca_by_category(pca_df, 'ConvolutionKernel', ax3, 'PCA by kernel', custom_colors_man, show_y_labels=False)
#ax3.set_ylabel('')  # Remove y-label for the second subplot to avoid redundancy

# Adjust layout and save the figure
plt.tight_layout()
plt.savefig(os.path.join(main_dir, 'results/PCA_Comparison_before_combat.png'))
plt.show()
```
