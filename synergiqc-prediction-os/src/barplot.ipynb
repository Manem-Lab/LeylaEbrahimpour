```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the best model c-index values
categories = ['Tumor', 'Non-Tumor'
              #, 'Delta'
             ]
models = ['Rads', 'Rads + Clinical']

# C-index values for each category and model
c_index_values = {
    'Tumor': [0.63, 0.64],
    'Non-Tumor': [0.60, 0.61]
#    ,'Delta': [, ]
}
# Confidence intervals (95%) for each category and model
conf_intervals = {
    'Tumor': [(0.03, 0.04), (0.03, 0.04)],
    'Non-Tumor': [(0.03, 0.06), (0.04, 0.05)]
}

# Set the width of the bars
bar_width = 0.35

# Create an array of indices for the x-axis
x = np.arange(len(categories))

# Create the figure and axes
fig, ax = plt.subplots()

# Plot the bars for each category and model
for i, model in enumerate(models):
    c_index_data = [c_index_values[category][i] for category in categories]
    conf_interval_data = [tuple(conf_intervals[category][i]) for category in categories]
    ax.bar(x + i * bar_width, c_index_data, bar_width, label=model, yerr=np.array(conf_interval_data).T, capsize=5)
    

# Set labels and title
ax.set_xlabel('Categories', fontsize=14)
ax.set_ylabel('C-Index', fontsize=14)
ax.set_title('Best Model C-Index for OS prediction')
ax.set_xticks(x + bar_width / 2)
ax.set_xticklabels(categories)
ax.legend()
# Set a custom y-axis limit
ax.set_ylim(0, 0.8)  # Adjust the limits as needed
ax.set_xlabel('Categories', labelpad=20)  # Increase label distance from the axis
ax.set_ylabel('C-Index', labelpad=20)  # Increase label distance from the axis

# Adjust layout
plt.tight_layout()

plt.savefig("/home/ulaval.ca/lesee/projects/Project2-synergiqc/plots/c-index-OS-st1_2.png")
# Display the plot
plt.show()
```
